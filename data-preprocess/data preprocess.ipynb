```python
import numpy as np

# Assume the paths to your .npy files are stored in lists
x_test_files = [
    '/content/drive/My Drive/Processed Data/digits_x_test_shunqi.npy',
    '/content/drive/My Drive/Processed Data/digits_x_test_carmen.npy',
    '/content/drive/My Drive/Processed Data/digits_x_test_dernice.npy',
    '/content/drive/My Drive/Processed Data/digits_x_test_shiting.npy',
    '/content/drive/My Drive/Processed Data/digits_x_test_ziaqian.npy'
]

y_test_files = [
    '/content/drive/My Drive/Processed Data/digits_y_test_shunqi.npy',
    '/content/drive/My Drive/Processed Data/digits_y_test_carmen.npy',
    '/content/drive/My Drive/Processed Data/digits_y_test_dernice.npy',
    '/content/drive/My Drive/Processed Data/digits_y_test_shiting.npy',
    '/content/drive/My Drive/Processed Data/digits_y_test_ziaqian.npy'
]

# Load the data
x_tests = [np.load(file) for file in x_test_files]
y_tests = [np.load(file) for file in y_test_files]

# Concatenate the arrays
x_test_combined = np.concatenate(x_tests, axis=0)
y_test_combined = np.concatenate(y_tests, axis=0)

# Save the concatenated arrays to .npy files
np.save('combined_x_test.npy', x_test_combined)
np.save('combined_y_test.npy', y_test_combined)

# Optionally, you can load them back to verify
combined_x_test = np.load('combined_x_test.npy')
combined_y_test = np.load('combined_y_test.npy')

print(f"x_test combined shape: {combined_x_test.shape}")
print(f"y_test combined shape: {combined_y_test.shape}")
```

```python
import matplotlib.pyplot as plt
def show(img, figsize=(8, 4), title=None):
    plt.figure(figsize=figsize)
    plt.imshow(img, cmap='gray')
    if title:
        plt.title(title)
    plt.show()
```
